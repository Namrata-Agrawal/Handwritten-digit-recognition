Import libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
```

Load the dataset and print the description

```python
digits=datasets.load_digits()
print(digits.DESCR)
```

```python
print(digits.data)
print(digits.target)
```

```python
plt.gray()
plt.matshow(digits.images[100])
#plt.show()
print("The digit at 100th position ",digits.target[100])
```

Training the model with dataset

```python
model=KMeans(n_clusters=10)
model.fit(digits.data)
```

Plot all the digits from 0-9

```python
fig = plt.figure(figsize=(8, 3))

print(fig.suptitle('Cluser Center Images', fontsize=14, fontweight='bold'))

for i in range(10):

  # Initialize subplots in a grid of 2X5, at i+1th position
  ax = fig.add_subplot(2, 5, 1 + i)

  # Display images
  ax.imshow(model.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)
plt.show()
```

```python
new_labels=model.predict(digits.data)
#print(new_labels)
for i in range(len(new_labels)):
    print(new_labels[i])
```

Full code implementation

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
digits=datasets.load_digits()
print(digits.DESCR)
plt.gray()
plt.matshow(digits.images[100])
plt.show()
print(digits.target[100])
model=KMeans(n_clusters=10)
model.fit(digits.data)
fig = plt.figure(figsize=(8, 3))

fig.suptitle('Cluser Center Images', fontsize=14, fontweight='bold')
for i in range(10):

  # Initialize subplots in a grid of 2X5, at i+1th position
  ax = fig.add_subplot(2, 5, 1 + i)

  # Display images
  ax.imshow(model.cluster_centers_[i].reshape((8, 8)), cmap=plt.cm.binary)
plt.show()
new_labels=model.predict(digits.data)
#print(new_labels)
for i in range(len(new_labels)):
    print(new_labels[i])
for i in range(len(new_labels)):
    if new_labels[i] == 0:
        print(0, end=' ')
    elif new_labels[i] == 1:
        print(9, end='')
    elif new_labels[i] == 2:
        print(2, end='')
    elif new_labels[i] == 3:
        print(1, end='')
    elif new_labels[i] == 4:
        print(6, end='')
    elif new_labels[i] == 5:
        print(8, end='')
    elif new_labels[i] == 6:
        print(4, end='')
    elif new_labels[i] == 7:
        print(5, end='')
    elif new_labels[i] == 8:
        print(7, end='')
    elif new_labels[i] == 9:
        print(3, end='')
```
